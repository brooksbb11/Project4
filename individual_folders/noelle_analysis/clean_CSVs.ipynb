### Blood Levels Testing Data

```python
# Read the Blood levels data into a Pandas DataFrame
blood_levels = pd.read_csv('https://health.data.ny.gov/resource/dyed-4zxh.csv')
blood_levels.head()
```

```python
# Get a brief summary of the Blood levels DataFrame.
blood_levels.info()
```

```python
#drop nulls
blood_levels = blood_levels.dropna()
blood_levels.info()
```

```python
# Get the blood_levels column names.
columns = list(blood_levels.columns)
print(columns)
```

```python
#Get the unique value counts in the columns
for i in columns: 
    print(i, blood_levels[i].nunique())
```

```python
#drop duplicates 
blood_levels = blood_levels.drop_duplicates(keep='first')
blood_levels.info()
```

```python
#Separate data for year 2020
blood_levels_2020 = blood_levels[blood_levels['year']== 2020]
blood_levels_2020.info()
```

```python
#Remove unwanted columns
blood_levels_2020 = blood_levels_2020.drop(columns=['year', 'countycode'])
blood_levels_2020.head()
```

```python
#sort and reset index
blood_levels_2020 = blood_levels_2020.sort_values(by=['county', 'zip', 'year_of_birth'])
blood_levels_2020 = blood_levels_2020.reset_index(drop=True)
blood_levels_2020.head(20)
```

```python
# Export the cleaned DataFrame as a CSV file. 
blood_levels_2020.to_csv("../../data/cleaned_blood_data.csv", encoding='utf8', index=False)
```

### School Water Testing Data

```python
# Read the School Water Testing data into a Pandas DataFrame
school_water = pd.read_csv('https://health.data.ny.gov/resource/4n6n-zu56.csv')
school_water.head()
```

```python
# Get a brief summary of the School Water DataFrame.
school_water.info()
```

```python
# Get the hosp_df column names.
school_columns = list(school_water.columns)
print(school_columns)
```

```python
#Get the unique value counts in the columns
for i in school_columns: 
    print(i, school_water[i].nunique())
```

```python
# drop Location column as it is empty
school_water = school_water.drop(columns=['location'])
school_water.info()
```

```python
#Drop nulls
school_water = school_water.dropna()
school_water.info()
```

```python
#drop unwanted columns
school_water = school_water.drop(columns=['compliance_year', 'school_district', 'sampling_complete', 'sampling_completion_date', 'remediation_status', 
                                          'school_website', 'out_of_service_or_addressed', 'all_results_received', 'date_all_results_received','beds_code', 
                                          'date_survey_updated'])
school_water.info()
```

```python
#drop duplicates
school_water = school_water.drop_duplicates(keep='first')
school_water.info()
```

```python
#rename columns
school_water.rename(columns={'number_of_outlets_result' : 'Number of Outlets, Result â‰¤ 15 ppb',
                             'number_of_outlets_result_1' :'Number of Outlets, Result > 15 ppb', 
                             'school_zip_code': 'Zip'}, inplace=True)
school_water.head()
```

```python
#sort and reset index
school_water = school_water.sort_values(by=['county', 'Zip', 'school'])
school_water = school_water.reset_index(drop=True)
school_water.head()
```

```python
#Reorganize columns

school_water = school_water[['county', 'Zip','School', 'Type of Organization', 'Any Building with Lead-Free Plumbing?', 'Results â‰¤ 15 ppb', 'Results â‰¤ 15 ppb', 'County Location']]
school_water.head()
```

```python
school_water.info()
```

```python
#Chaning dtypes
school_water = school_water.astype({'Zip': 'int64'})
school_water.info()
```

```python
# Export the cleaned DataFrame as a CSV file. 
school_water.to_csv("../../data/cleaned_water_data.csv", encoding='utf8', index=False)
```
