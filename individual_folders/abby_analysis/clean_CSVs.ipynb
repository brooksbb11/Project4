```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

### Extract the project 4 Data

### Blood Levels Testing Data

```python
# Read the Blood levels data into a Pandas DataFrame
blood_levels = pd.read_csv('../../data/Childhood_Blood_Lead_Testing.csv')
blood_levels.head()
```

```python
# Get a brief summary of the Blood levels DataFrame.
blood_levels.info()
```

```python
#drop nulls
blood_levels = blood_levels.dropna()
blood_levels.info()
```

```python
# Get the blood_levels column names.
columns = list(blood_levels.columns)
print(columns)
```

```python
#Get the unique value counts in the columns
for i in columns: 
    print(i, blood_levels[i].nunique())
```

```python
#drop duplicates 
blood_levels = blood_levels.drop_duplicates(keep='first')
blood_levels.info()
```

```python
#Separate data for year 2020
blood_levels_2020 = blood_levels[blood_levels['Year']== 2020]
blood_levels_2020.info()
```

```python
#Remove unwanted columns
blood_levels_2020 = blood_levels_2020.drop(columns=['Year', 'County Code'])
blood_levels_2020.head()
```

```python
#sort and reset index
blood_levels_2020 = blood_levels_2020.sort_values(by=['County', 'Zip', 'Year of Birth'])
blood_levels_2020 = blood_levels_2020.reset_index(drop=True)
blood_levels_2020.head(20)
```

```python
# Export the cleaned DataFrame as a CSV file. 
blood_levels_2020.to_csv("../../data/cleaned_blood_data.csv", encoding='utf8', index=False)
```

### School Water Testing Data

```python
# Read the School Water Testing data into a Pandas DataFrame
school_water = pd.read_csv('../../data/Lead_Testing_in_School_Drinking_Water_Sampling.csv')
school_water.head()
```

```python
# Get a brief summary of the School Water DataFrame.
school_water.info()
```

```python
# Get the hosp_df column names.
school_columns = list(school_water.columns)
print(school_columns)
```

```python
#Get the unique value counts in the columns
for i in school_columns: 
    print(i, school_water[i].nunique())
```

```python
# drop Location column as it is empty
school_water = school_water.drop(columns=['Location'])
school_water.info()
```

```python
#Drop nulls
school_water = school_water.dropna()
school_water.info()
```

```python
#drop unwanted columns
school_water = school_water.drop(columns=['Compliance Year', 'School District', 'Sampling Complete', 'Sampling Completion Date', 'Remediation Status', 
                                          'School Website', 'Out of Service or Addressed', 'All Results Received', 'Date All Results Received','BEDS Code', 
                                          'School Street', 'School City', 'School State','Date Survey Updated'])
school_water.info()
```

```python
#drop duplicates
school_water = school_water.drop_duplicates(keep='first')
school_water.info()
```

```python
#rename columns
school_water.rename(columns={'Number of Outlets, Result ≤ 15 ppb': 'Results ≤ 15 ppb', 'Number of Outlets, Result > 15 ppb': 'Results ≤ 15 ppb', 
                             'School ZIP Code': 'Zip'}, inplace=True)
school_water.head()
```

```python
#sort and reset index
school_water = school_water.sort_values(by=['County', 'Zip', 'School'])
school_water = school_water.reset_index(drop=True)
school_water.head()
```

```python
#Reorganize columns
school_water = school_water[['County', 'Zip','School', 'Type of Organization', 'Any Building with Lead-Free Plumbing?', 'Number of Outlets that Require Sampling', 'Results ≤ 15 ppb', 'Results ≤ 15 ppb', 'County Location']]
school_water.head()
```

```python
school_water.info()
```

```python
#Chaning dtypes
school_water = school_water.astype({'Zip': 'int64'})
school_water.info()
```

```python
# Export the cleaned DataFrame as a CSV file. 
school_water.to_csv("../../data/cleaned_water_data.csv", encoding='utf8', index=False)
```

### SVI Data

```python
# Read the SVI data into a Pandas DataFrame
svi_df = pd.read_csv('../../data/NewYork_county_SVI.csv')
svi_df.head()
```

```python
# Get a brief summary of the No Heat days DataFrame.
svi_df.info()
```

```python
#Get the no_heat_days_df column names.
svi_columns = list(svi_df.columns)
print(svi_columns)
```

```python
#Get the unique value counts in the columns
for i in svi_columns: 
    print(i, svi_df[i].nunique())
```

```python
#Pull desired columns
e_series = svi_df.loc[:, svi_df.columns.str.startswith('E_')]
ep_series = svi_df.loc[:, svi_df.columns.str.startswith('EP_')]
rpl_series = svi_df.loc[:, svi_df.columns.str.startswith('RPL_')]
```

```python
svi_des_columns = []
svi_des_columns.extend(list(e_series.columns))
svi_des_columns.extend(list(ep_series.columns))
svi_des_columns.extend(list(rpl_series.columns))
```

```python
#Merge the columns to a single dataframe.
svi = svi_df[svi_des_columns]
svi['COUNTY'] = svi_df['COUNTY']
svi.head()
```

```python
#Reorganize columns
svi = svi[['COUNTY', 'E_TOTPOP', 'E_HU', 'E_HH', 'E_POV150', 'E_UNEMP', 'E_HBURD',
       'E_NOHSDP', 'E_UNINSUR', 'E_AGE65', 'E_AGE17', 'E_DISABL', 'E_SNGPNT',
       'E_LIMENG', 'E_MINRTY', 'E_MUNIT', 'E_MOBILE', 'E_CROWD', 'E_NOVEH',
       'E_GROUPQ', 'E_DAYPOP', 'E_NOINT', 'E_AFAM', 'E_HISP', 'E_ASIAN',
       'E_AIAN', 'E_NHPI', 'E_TWOMORE', 'E_OTHERRACE', 'EP_POV150', 'EP_UNEMP',
       'EP_HBURD', 'EP_NOHSDP', 'EP_UNINSUR', 'EP_AGE65', 'EP_AGE17',
       'EP_DISABL', 'EP_SNGPNT', 'EP_LIMENG', 'EP_MINRTY', 'EP_MUNIT',
       'EP_MOBILE', 'EP_CROWD', 'EP_NOVEH', 'EP_GROUPQ', 'EP_NOINT', 'EP_AFAM',
       'EP_HISP', 'EP_ASIAN', 'EP_AIAN', 'EP_NHPI', 'EP_TWOMORE',
       'EP_OTHERRACE', 'RPL_THEME1', 'RPL_THEME2', 'RPL_THEME3', 'RPL_THEME4',
       'RPL_THEMES']]

svi.info()
```

```python
#sort and reset index
svi = svi.sort_values(by=['COUNTY'])
svi = svi.reset_index(drop=True)
svi.head()
```

```python
# Export the cleaned DataFrame as a CSV file. 
svi.to_csv("../../data/cleaned_svi_data.csv", encoding='utf8', index=False)
```
